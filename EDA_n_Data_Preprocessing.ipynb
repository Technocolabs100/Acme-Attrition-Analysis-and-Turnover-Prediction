```python
# Function to plot pie chart.

def plot_pie(column):
    target = dataset[column].value_counts()    
    fig1, ax1 = plt.subplots()    
    ax1.pie(target, labels=target.index, autopct='%1.1f%%', shadow=None)
    ax1.axis('equal')
    plt.title("All Employees", fontsize=14)
    plt.show()
    

def plot_compare(column, category_name):
    NChurned = len(attrited[column].unique())
    NNonChurned = len(nonattrited[column].unique())
    
    ChurnedCounts = attrited[column].value_counts().sort_index()
    NonChurnedCounts = nonattrited[column].value_counts().sort_index()
    
    indchurned = np.arange(NChurned)    # the x locations for the groups
    indnonchurned = np.arange(NNonChurned)    
    width = 1       # the width of the bars: can also be len(x) sequence
    
    figs, axs = plt.subplots(1,2, figsize=(12,5))
    
    axs[1].bar(indchurned, ChurnedCounts, width, color='#DD8452')
    axs[1].set_title('Churned ' + category_name, fontsize=20)
    axs[1].set_xticks(indchurned)
    axs[1].set_xticklabels(ChurnedCounts.index.tolist(), rotation=45)
    
    axs[0].bar(indnonchurned, NonChurnedCounts, width, color='b')
    axs[0].set_title('Non Churned ' + category_name, fontsize=20)
    axs[0].set_ylabel('Amount of People')
    axs[0].set_xticks(indnonchurned)
    axs[0].set_xticklabels(NonChurnedCounts.index.tolist(), rotation=45)
    
    
    plt.show()
```

# Correlation.

```python
# Replace 'Yes' with 1 and 'No' with 0 in the 'response' column
dataset['Attrition'] = dataset['Attrition'].replace({'Yes': 1, 'No': 0})
```

```python
corr_features = continuous_features.copy()
corr_features.append('Attrition')
print(corr_features)
```

```python
cont_num_df = dataset[continuous_features]
cont_num_df['Attrition'] = dataset['Attrition']
```

```python
cont_num_df.head()
```

```python
# Calculate the correlation matrix
corr_matrix = cont_num_df.corr().abs()


# corr_mat = dataset[corr_features].corr()
# #top_corr_features = corr_mat.index

plt.figure(figsize=(20,20))

# Heat Map
sns.heatmap(corr_matrix, annot=True, cmap="RdYlGn")
```

Remove highly correlated features to avoid multicollinearity problem.

```python
# Create an upper triangle matrix to avoid duplicate pairs
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Set a threshold value for dropping correlated features
threshold = 0.7

# Step 5: Find the columns that have correlations above the threshold
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]

print(to_drop)

# Step 6: Drop the correlated features
reduced_df = cont_num_df.drop(columns=to_drop)

# Show the remaining features
print("Dropped features:", to_drop)
print("Remaining features after dropping correlated ones:")
print(reduced_df.head())
```

The features 'TotalWorkingYears', 'YearsInCurrentRole' and 'YearsWithCurrManager' are highly correlated to each other. So, it has been dropped to avoid multicollinearity problem.

ENCODING

In our dataset: "BusinessTravel, Department, EducationField, Gender, JobRole, MaritalStatus, Over18 and OverTime" are the categorical features. All these features are nominal categorical features. So, One-hot encoding is performed.

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']

categorical_features
```

```python
# Number of categories each categorical feature has:

for feature in categorical_features:
    print("For feature {}, the cardinality is {}".format(feature, len(dataset[feature].unique())))
```

```python
# Performing One-Hot Encoding 

dataset_encoded = pd.get_dummies(dataset, columns=['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime'])
```

```python
#print(dataset_encoded)

dataset_encoded.head()
```

```python
dataset_encoded = dataset_encoded.replace({True: 1, False: 0})
dataset_encoded.head()
```

```python
dataset_encoded.shape
```

# FEATURE SCALING

```python
feature_scale = [feature for feature in dataset_encoded.columns if feature not in ['EmployeeNumber', 'Attrition']]
```

```python
# Standard Scaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(dataset_encoded[feature_scale])
```

```python
preProcessed_data = pd.concat([dataset_encoded[['EmployeeNumber', 'Attrition']].reset_index(drop=True),
                               pd.DataFrame(scaler.transform(dataset_encoded[feature_scale]), columns = feature_scale)],
                               axis=1)
```

```python
preProcessed_data.head()
```

# Feature Selection

Dropping constant features - Variance threshold

```python
# Removing 0 variance features

preProcessed_data.head()
```

```python
from sklearn.feature_selection import VarianceThreshold

var_thres = VarianceThreshold(threshold = 0)
var_thres.fit(preProcessed_data)
```

```python
# True - Columns containing non 0 variance, False - Columns containing 0 variance.

var_thres.get_support()
```

```python
const_columns = [column for column in preProcessed_data.columns if column not in preProcessed_data.columns[var_thres.get_support()]]

print(const_columns)
```

```python
# Deleting columns with 0 variance

dataset_final = preProcessed_data.drop(const_columns, axis=1)
```
